```python
import pandas as pd
import re
import ast

# ------------------------
# 1. Charger le fichier
# ------------------------
df = pd.read_csv(r"C:\Users\TCHUI\Downloads\products.csv")

# ------------------------
# 2. Nettoyage des colonnes
# ------------------------
# uniformiser les noms des colonnes
df.columns = df.columns.str.strip().str.replace('/', '_').str.replace(' ', '_').str.lower()

# gérer les valeurs manquantes
df['customer_review_summary'] = df['customer_review_summary'].fillna('No review')
df['recent_purchases'] = df['recent_purchases'].fillna('0')
df['default_variant_0'] = df['default_variant_0'].fillna('')


# ------------------------
# 3. Extraire les données numériques et catégorielles
# ------------------------
# Best sellers rank (extraction des chiffres)
df['best_sellers_rank'] = df['best_sellers_rank'].str.extract(r'#([\d,]+)')[0]
df['best_sellers_rank'] = df['best_sellers_rank'].str.replace(',', '').astype(float)

# Note moyenne
df['rating'] = df['rating_stars'].str.extract(r'([0-9.]+)').astype(float)

# Taille du produit
df['size'] = df['default_variant_0'].str.extract(r'size:([A-Za-z0-9-]+)')

# Nombre d'achats récents
def extract_number(x):
    if pd.isna(x):
        return 0
    match = re.search(r'(\d+)', str(x).replace('+',''))
    return int(match.group(1)) if match else 0

df['recent_purchases_count'] = df['recent_purchases'].apply(extract_number)

# Catégorie principale
df['main_category'] = df['breadcrumbs'].str.split('›').str[0].str.strip()

# Nombre d'images
def count_images(x):
    try:
        imgs = ast.literal_eval(x)
        return len(imgs)
    except:
        return 0

df['image_count'] = df['all_images'].apply(count_images)

# Longueur de la description
df['desc_length'] = df['about_item'].apply(lambda x: len(str(x).split()))

# Score global (rating x popularité)
df['popularity_score'] = df['rating'] * (1 + df['recent_purchases_count']/100)

# ------------------------
# 4. Sauvegarder le dataset enrichi
# ------------------------
df.to_csv(r"C:\Users\TCHUI\Downloads\products_enriched.csv", index=False)

print("Dataset enrichi enregistré avec succès !")
print(df.head(10))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# ------------------------
# 1️⃣ Charger le dataset enrichi
# ------------------------
df = pd.read_csv(r"C:\Users\TCHUI\Downloads\products_enriched.csv")

# ------------------------
# 2️⃣ Préparer les features et la cible
# ------------------------
# Colonnes numériques pertinentes pour la prédiction
features = ['best_sellers_rank', 'recent_purchases_count', 'desc_length', 'image_count']
X = df[features].fillna(0)

# Cible : rating
y = df['rating'].fillna(df['rating'].mean())

# Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Séparer en train/test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ------------------------
# 3️⃣ Construire le modèle
# ------------------------
from tensorflow.keras.layers import Input

model = Sequential([
    Input(shape=(X_train.shape[1],)),  # couche d'entrée correcte
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1)  # sortie pour régression
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

# ------------------------
# 4️⃣ Entraîner le modèle
# ------------------------
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# ------------------------
# 5️⃣ Évaluer le modèle
# ------------------------
loss, mae = model.evaluate(X_test, y_test)
print(f"MAE sur le test set: {mae:.2f}")

# ------------------------
# 6️⃣ Faire des prédictions
# ------------------------
y_pred = model.predict(X_test)
print("Exemple de prédictions :", y_pred[:10])
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# y_test : vraies notes
# y_pred : prédictions du modèle (prédites précédemment)

# Si y_pred est un tableau 2D de forme (n,1), on le convertit en 1D
y_pred_flat = y_pred.flatten()

# 1️⃣ Calcul des métriques
mae = mean_absolute_error(y_test, y_pred_flat)
mse = mean_squared_error(y_test, y_pred_flat)
r2 = r2_score(y_test, y_pred_flat)

print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
print(f"R²: {r2:.3f}")

# 2️⃣ Visualisation : vraies notes vs prédictions
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred_flat, alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("Vraies notes")
plt.ylabel("Prédictions")
plt.title("Prédictions vs Vraies notes")
plt.show()

# 3️⃣ Optionnel : visualiser les erreurs
errors = y_test - y_pred_flat
plt.figure(figsize=(8,6))
plt.hist(errors, bins=20, color='skyblue', edgecolor='black')
plt.xlabel("Erreur")
plt.ylabel("Nombre de produits")
plt.title("Distribution des erreurs de prédiction")
plt.show()
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, Concatenate, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# ------------------------
# 1️⃣ Charger le dataset
# ------------------------
df = pd.read_csv(r"C:\Users\TCHUI\Downloads\products_enriched.csv")

# Remplir les valeurs manquantes
df['rating'] = df['rating'].fillna(df['rating'].mean())
df['title'] = df['title'].fillna("")

# ------------------------
# 2️⃣ Préparer les features numériques
# ------------------------
num_features = ['best_sellers_rank', 'recent_purchases_count', 'desc_length', 'image_count']
X_num = df[num_features].fillna(0)
scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(X_num)

# ------------------------
# 3️⃣ Préparer les features textuelles
# ------------------------
MAX_WORDS = 5000
MAX_LEN = 20

tokenizer = Tokenizer(num_words=MAX_WORDS)
tokenizer.fit_on_texts(df['title'])
X_text_seq = tokenizer.texts_to_sequences(df['title'])
X_text_pad = pad_sequences(X_text_seq, maxlen=MAX_LEN, padding='post')

# ------------------------
# 4️⃣ Cible
# ------------------------
y = df['rating'].values

# ------------------------
# 5️⃣ Split train/test
# ------------------------
X_num_train, X_num_test, X_text_train, X_text_test, y_train, y_test = train_test_split(
    X_num_scaled, X_text_pad, y, test_size=0.2, random_state=42
)

# ------------------------
# 6️⃣ Construire le modèle
# ------------------------
# Branche numérique
num_input = Input(shape=(X_num_train.shape[1],))
x_num = Dense(32, activation='relu')(num_input)
x_num = Dropout(0.2)(x_num)

# Branche texte
text_input = Input(shape=(MAX_LEN,))
x_text = Embedding(input_dim=MAX_WORDS, output_dim=32, input_length=MAX_LEN)(text_input)
x_text = LSTM(32)(x_text)

# Fusion des branches
x = Concatenate()([x_num, x_text])
x = Dense(32, activation='relu')(x)
output = Dense(1)(x)

model = Model(inputs=[num_input, text_input], outputs=output)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

# ------------------------
# 7️⃣ Entraînement
# ------------------------
history = model.fit(
    [X_num_train, X_text_train],
    y_train,
    epochs=30,
    batch_size=32,
    validation_split=0.2
)

# ------------------------
# 8️⃣ Évaluation
# ------------------------
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict([X_num_test, X_text_test]).flatten()

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.3f}, MSE: {mse:.3f}, R²: {r2:.3f}")
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from transformers import BertTokenizer, TFBertModel
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# ------------------------
# 1️⃣ Charger le dataset
# ------------------------
df = pd.read_csv(r"C:\Users\TCHUI\Downloads\products_enriched.csv")
df['rating'] = df['rating'].fillna(df['rating'].mean())
df['title'] = df['title'].fillna("")

# ------------------------
# 2️⃣ Features numériques
# ------------------------
num_features = ['best_sellers_rank', 'recent_purchases_count', 'desc_length', 'image_count']
X_num = df[num_features].fillna(0)
scaler = StandardScaler()
X_num_scaled = scaler.fit_transform(X_num)

# ------------------------
# 3️⃣ Texte avec BERT
# ------------------------
MAX_LEN = 32
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
tokens = tokenizer(
    list(df['title']),
    max_length=MAX_LEN,
    padding='max_length',
    truncation=True,
    return_tensors='tf'
)

input_ids = tokens['input_ids']
attention_mask = tokens['attention_mask']

# ------------------------
# 4️⃣ Cible
# ------------------------
y = df['rating'].values

# ------------------------
# 5️⃣ Split train/test
# ------------------------
X_num_train, X_num_test, input_ids_train, input_ids_test, mask_train, mask_test, y_train, y_test = train_test_split(
    X_num_scaled, input_ids, attention_mask, y, test_size=0.2, random_state=42
)

# ------------------------
# 6️⃣ Construire le modèle
# ------------------------
# Inputs
num_input = tf.keras.layers.Input(shape=(X_num_train.shape[1],), name="numerical_input")
ids_input = tf.keras.layers.Input(shape=(MAX_LEN,), dtype=tf.int32, name="input_ids")
mask_input = tf.keras.layers.Input(shape=(MAX_LEN,), dtype=tf.int32, name="attention_mask")

# BERT pour texte
bert_model = TFBertModel.from_pretrained('bert-base-uncased')
bert_output = bert_model(input_ids=ids_input, attention_mask=mask_input)[1]  # pooled output

# Dense pour features numériques
x_num = tf.keras.layers.Dense(32, activation='relu')(num_input)

# Fusion
x = tf.keras.layers.Concatenate()([bert_output, x_num])
x = tf.keras.layers.Dense(64, activation='relu')(x)
x = tf.keras.layers.Dropout(0.2)(x)
output = tf.keras.layers.Dense(1)(x)

model = tf.keras.Model(inputs=[num_input, ids_input, mask_input], outputs=output)
model.compile(optimizer=tf.keras.optimizers.Adam(1e-5), loss='mse', metrics=['mae'])
model.summary()

# ------------------------
# 7️⃣ Entraînement
# ------------------------
history = model.fit(
    [X_num_train, input_ids_train, mask_train],
    y_train,
    validation_split=0.2,
    epochs=5,
    batch_size=16
)

# ------------------------
# 8️⃣ Évaluation
# ------------------------
y_pred = model.predict([X_num_test, input_ids_test, mask_test]).flatten()

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.3f}, MSE: {mse:.3f}, R²: {r2:.3f}")
```
