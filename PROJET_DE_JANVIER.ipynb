```python
!pip install librosa
import librosa
import numpy as np

def extraire_features(fichier_audio):
    y, sr = librosa.load(fichier_audio, sr=16000)
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)
    return np.mean(mfcc.T, axis=0)
```
